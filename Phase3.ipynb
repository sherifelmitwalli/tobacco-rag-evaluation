```python
!pip install pypdf
!pip install openai==0.28
!pip install chromadb
!pip install typing
!pip install sentence-transformers
!pip install transformers
!pip install deepinfra[openai]

import os
import re
import csv
import time
import random
import openai
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
from typing import List, Tuple
from pypdf import PdfReader
import chromadb
from sentence_transformers import SentenceTransformer
from chromadb.api.types import Documents, Embeddings, EmbeddingFunction
from transformers import AutoTokenizer, AutoModelForCausalLM
from huggingface_hub import login
from itertools import product

# Load the PDF file and extract text from each page
def load_pdf(file_path: str) -> str:
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text
    return text

# Split the text into chunks

def split_text(text: str, chunk_size: int, overlap: int) -> List[str]:
    words = text.split()
    chunks = []
    overlap_words = int(chunk_size * (overlap / 100))

    for i in range(0, len(words), chunk_size - overlap_words):
        chunk = " ".join(words[i:i + chunk_size])
        chunks.append(chunk)
    return chunks

# Define a custom embedding function using SentenceTransformer
class CustomEmbeddingFunction(EmbeddingFunction):
    def __init__(self, model_name: str):
        self.model = SentenceTransformer(model_name)

    def __call__(self, input: Documents) -> Embeddings:
        return self.model.encode(input).tolist()

# Create and populate Chroma database
def create_chroma_db(documents: List[str], name: str, embedding_model: str) -> chromadb.Collection:
    db_folder = "chroma_db"
    if not os.path.exists(db_folder):
        os.makedirs(db_folder)

    chroma_client = chromadb.PersistentClient(path=db_folder)

    existing_collections = chroma_client.list_collections()
    if name in [c.name for c in existing_collections]:
        chroma_client.delete_collection(name)
        print(f"Deleted existing collection: {name}")

    embedding_function = CustomEmbeddingFunction(embedding_model)
    db = chroma_client.create_collection(name=name, embedding_function=embedding_function)
    for i, d in enumerate(documents):
        db.add(documents=[d], ids=[str(i)])

    return db

# Retrieve the most relevant passages based on the query
def get_relevant_passage(query: str, db: chromadb.Collection, n_results: int) -> List[str]:
    results = db.query(query_texts=[query], n_results=n_results)
    return results['documents'][0]

# Function to generate response using DeepInfra
# Function to generate response using DeepInfra
def generate_deepinfra_response(query: str, context: str, model_name: str, max_input_length=32767) -> str:
    prompt = f"Answer the following question based on the given context. Question: {query} Context: {context}"

    # Truncate the context if it exceeds the maximum input length
    if len(prompt) > max_input_length:
        prompt = prompt[:max_input_length]
    openai.api_key = "deepinfra_api_key"
    openai.api_base = "https://api.deepinfra.com/v1/openai"

    chat_completion = openai.ChatCompletion.create(
        model="mistralai/Mixtral-8x7B-Instruct-v0.1", # Changed model name to a valid model on DeepInfra
        messages=[{"role": "user", "content": prompt}],
    )
    return chat_completion.choices[0].message.content

# Function to read queries from CSV
def read_queries_from_csv(file_path):
    queries = []
    with open(file_path, 'r', newline='', encoding='utf-8-sig') as file:
        reader = csv.DictReader(file)
        for i, fieldname in enumerate(reader.fieldnames):
            if fieldname.startswith('\ufeff'):
                reader.fieldnames[i] = fieldname[len('\ufeff'):]
        if 'Query' not in reader.fieldnames:
            print(f"Error: 'Query' column not found in CSV. Available columns: {reader.fieldnames}")
            raise KeyError(f"'Query' column not found in CSV. Available columns: {reader.fieldnames}")
        for row in reader:
            queries.append(row['Query'])
    return queries

# Function to evaluate response using GPT-4o
def evaluate_response(query: str, response: str) -> List[float]:
    max_retries = 5
    base_delay = 60  # 60 seconds
    # Setting OpenAI API key and base URL for evaluation
    openai.api_key = "openai_api_key" # Replace with your OpenAI API key
    openai.api_base = "https://api.openai.com/v1" # This is the default for OpenAI's API

    for attempt in range(max_retries):
        try:
            evaluation_prompt = f"""
            Evaluate the following response to the query based on the given criteria:

            Query: {query}
            Response: {response}

            Criteria:
            1. Relevance (0-10): How well does the response address the specific query?
            2. Accuracy (0-10): How factually correct and precise is the response?
            3. Completeness (0-10): Does the response provide a comprehensive answer covering all aspects of the query?
            4. Clarity (0-10): Is the response clear and easy to understand?
            5. Conciseness (0-10): Is the response brief yet informative, avoiding unnecessary details?
            6. Coherence (0-10): Is the response logically structured and coherent?
            7. Context Awareness (0-10): Does the response demonstrate understanding of the context within the document?
            8. Adaptability (0-10): How well does the response adapt to the nuances of the query?
            9. Domain Specificity (0-10): Does the response show specialized knowledge relevant to tobacco?

            Provide only the scores for each criterion, separated by commas, in the order listed above. Do not include any additional text, explanations, or the criteria numbers.
            """
            response = openai.ChatCompletion.create(
                model="gpt-4o",
                messages=[
                    {"role": "system", "content": "You are an evaluator that provides detailed feedback on query responses."},
                    {"role": "user", "content": evaluation_prompt}
                ],
                max_tokens=100
            )

            # Clean and process the response
            content = response['choices'][0]['message']['content']
            scores_str = re.findall(r'\d+(?:\.\d+)?', content)

            if len(scores_str) != 9:
                raise ValueError(f"Expected 9 scores, but got {len(scores_str)}")

            scores = [float(score) for score in scores_str]
            return scores
        except (openai.error.RateLimitError, openai.error.APIError) as e:
            if attempt < max_retries - 1:
                delay = base_delay * (2 ** attempt) + random.uniform(0, 10)
                print(f"Error occurred: {e}. Retrying in {delay:.2f} seconds...")
                time.sleep(delay)
            else:
                raise
        except ValueError as ve:
            print(f"Error processing scores: {ve}")
            print(f"Raw content: {content}")
            raise

# Grid search function
def grid_search(pdf_text: str, queries: List[str], model_name: str) -> Tuple[Tuple, List[Tuple]]:

    # Define hyperparameter grid
    chunk_sizes = [100, 300, 500]
    overlaps = [0, 25, 50]
    embedding_models = ['all-MiniLM-L6-v2', 'paraphrase-MiniLM-L6-v2']
    n_results_list = [1, 3, 5]

    best_score = 0
    best_params = None
    results = []

    for chunk_size, overlap, embedding_model, n_results in product(chunk_sizes, overlaps, embedding_models, n_results_list):
        print(f"Testing parameters: chunk_size={chunk_size}, overlap={overlap}, embedding_model={embedding_model}, n_results={n_results}")

        # Create chunks and database
        chunks = split_text(pdf_text, chunk_size, overlap)
        db = create_chroma_db(chunks, f"tobacco_report_{chunk_size}_{overlap}_{embedding_model}", embedding_model)

        total_scores = [0] * 9  # 9 criteria
        for query in queries:
            relevant_passages = get_relevant_passage(query, db, n_results)
            context = " ".join(relevant_passages)
            response = generate_deepinfra_response(query, context, model_name)
            scores = evaluate_response(query, response)
            total_scores = [t + s for t, s in zip(total_scores, scores)]

        avg_scores = [score / len(queries) for score in total_scores]
        overall_score = sum(avg_scores) / len(avg_scores)
        results.append((overall_score, avg_scores, chunk_size, overlap, embedding_model, n_results))
        print(f"Average scores: {avg_scores}")
        print(f"Overall score: {overall_score}")

        if overall_score > best_score:
            best_score = overall_score
            best_params = (chunk_size, overlap, embedding_model, n_results)

    results.sort(key=lambda x: x[0], reverse=True)  # Sort by overall score
    return best_params, results

# Function to plot results
def plot_results(results: List[Tuple], final_evaluations: List[List[float]]):
    df = pd.DataFrame(results, columns=['Overall Score', 'Avg Scores', 'Chunk Size', 'Overlap', 'Embedding Model', 'N Results'])

    # Original plots
    fig, axs = plt.subplots(2, 2, figsize=(20, 20))

    axs[0, 0].scatter(df['Chunk Size'], df['Overall Score'])
    axs[0, 0].set_xlabel('Chunk Size')
    axs[0, 0].set_ylabel('Overall Score')
    axs[0, 0].set_title('Chunk Size vs Overall Score')

    axs[0, 1].scatter(df['Overlap'], df['Overall Score'])
    axs[0, 1].set_xlabel('Overlap')
    axs[0, 1].set_ylabel('Overall Score')
    axs[0, 1].set_title('Overlap vs Overall Score')

    axs[1, 0].scatter(df['N Results'], df['Overall Score'])
    axs[1, 0].set_xlabel('N Results')
    axs[1, 0].set_ylabel('Overall Score')
    axs[1, 0].set_title('N Results vs Overall Score')

    embedding_models = df['Embedding Model'].unique()
    scores = [df[df['Embedding Model'] == model]['Overall Score'].mean() for model in embedding_models]
    axs[1, 1].bar(embedding_models, scores)
    axs[1, 1].set_xlabel('Embedding Model')
    axs[1, 1].set_ylabel('Average Overall Score')
    axs[1, 1].set_title('Embedding Model vs Average Overall Score')

    plt.tight_layout()
    plt.savefig('grid_search_results.png')
    plt.show()  # Display the radar plot

    plt.close()

    # Radar plot
    metrics = ['Relevance', 'Accuracy', 'Completeness', 'Clarity', 'Conciseness',
               'Coherence', 'Context Awareness', 'Adaptability', 'Domain Specificity']

    avg_scores = np.mean(final_evaluations, axis=0)

    angles = np.linspace(0, 2*np.pi, len(metrics), endpoint=False)
    avg_scores = np.concatenate((avg_scores, [avg_scores[0]]))  # repeat the first value to close the polygon
    angles = np.concatenate((angles, [angles[0]]))  # repeat the first angle to close the polygon

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))
    ax.plot(angles, avg_scores)
    ax.fill(angles, avg_scores, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metrics)
    ax.set_ylim(0, 10)
    ax.set_yticks(np.arange(2, 11, 2))
    ax.set_title("Performance Across Metrics")

    plt.tight_layout()
    plt.savefig('radar_plot.png')
    plt.show()  # Display the radar plot

    plt.close()

# Main execution
if __name__ == "__main__":

    # Set OpenAI API key and base URL for DeepInfra
    openai.api_key = "0Un56fWjpjAL4zCqyhpqYMbtq7EqrTpl"
    openai.api_base = "https://api.deepinfra.com/v1/openai"

    # Load and process the PDF
    pdf_path = "/content/pmi_report_2023.pdf"
    pdf_text = load_pdf(pdf_path)

    # File paths
    input_csv = 'query_response_reference1.csv'

    # Read queries from the input CSV
    queries = read_queries_from_csv(input_csv)

    # Set device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Perform grid search
    deepinfra_model_name = "mistralai/Mixtral-8x7B-Instruct-v0.1"
    try:
        best_params, all_results = grid_search(pdf_text, queries, deepinfra_model_name)
    except openai.error.InvalidRequestError as e:
        print(f"Error occurred: {e}")
        print("Please ensure that the model name is correct and available in the DeepInfra API.")
        print("You can check the available models by visiting the DeepInfra API documentation or contacting DeepInfra support.")
        raise

    print("Grid search completed.")
    print(f"Optimal hyperparameters:")
    print(f"Chunk size: {best_params[0]}")
    print(f"Overlap: {best_params[1]}")
    print(f"Embedding model: {best_params[2]}")
    print(f"Number of retrieved passages: {best_params[3]}")
    print(f"Best overall score: {all_results[0][0]}")

    # Save grid search results
    results_csv = 'grid_search_results.csv'
    with open(results_csv, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Overall Score", "Relevance", "Accuracy", "Completeness", "Clarity", "Conciseness", "Coherence", "Context Awareness", "Adaptability", "Domain Specificity", "Chunk Size", "Overlap", "Embedding Model", "N Results"])
        for result in all_results:
            writer.writerow([result[0]] + result[1] + list(result[2:]))

    print(f"Grid search results have been saved to {results_csv}")

    # Generate final responses using best parameters
    chunks = split_text(pdf_text, best_params[0], best_params[1])
    db = create_chroma_db(chunks, f"tobacco_report_best", best_params[2])

    final_responses = []
    final_evaluations = []
    for query in queries:
        relevant_passages = get_relevant_passage(query, db, best_params[3])
        context = " ".join(relevant_passages)
        response = generate_deepinfra_response(query, context, deepinfra_model_name)
        final_responses.append(response)

        # Set OpenAI API key and base URL for evaluation
        openai.api_key = "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"
        openai.api_base = "https://api.openai.com/v1"

        evaluation = evaluate_response(query, response)
        final_evaluations.append(evaluation)

    # Save final responses and evaluations
    output_csv = 'final_results.csv'
    with open(output_csv, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Query", "Response", "Relevance", "Accuracy", "Completeness", "Clarity", "Conciseness", "Coherence", "Context Awareness", "Adaptability", "Domain Specificity"])
        for query, response, evaluation in zip(queries, final_responses, final_evaluations):
            writer.writerow([query, response] + evaluation)

    print(f"Final results have been saved to {output_csv}")

    # Plot results
    plot_results(all_results, final_evaluations)

    print("Execution completed. Check the output CSV file and generated plots.")
```
